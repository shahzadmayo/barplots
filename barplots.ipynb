## Plot Bar graph using seaborn.barplot() method.

```python
# importing the required library
import seaborn as sns
import matplotlib.pyplot as plt
 
# read a titanic.csv file
# from seaborn library
kasti = sns.load_dataset('titanic')
 
# who v/s fare barplot
sns.barplot(x = 'who',
            y = 'fare',
            data = kasti)
 
# Show the plot
plt.show()
```

# Draw Barplot using Hue Function


```python
# importing the required library
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
 
# read a titanic.csv file
# from seaborn library
kasti1 = sns.load_dataset('titanic')
#kasti1.head()

sns.barplot(x = 'who',
            y = 'fare',
            hue = 'class',
            data = kasti1)
 
# who v/s fare barplot
plt.show()
# Show the plot
```

```python
# read a titanic.csv file
# from seaborn library
kasti = sns.load_dataset('titanic')
 
 
# who v/s fare barplot
sns.barplot(x = 'who',
            y = 'fare',
            hue = 'class',
            data = kasti,
            palette = "Blues")
 
# Show the plot
plt.show()
```

```python
# read a titanic.csv file
# from seaborn library
df = sns.load_dataset('titanic')
 
 
# who v/s fare barplot
sns.barplot(x = 'who',
            y = 'fare',
            hue = 'class',
            data = df,
            estimator = np.median,
            ci = 0)
 
# Show the plot
plt.show()
```

## Draw a set of vertical bar plots grouped by a categorical variable Creating a simple bar plot using seaborn

```python
import matplotlib.pyplot as plt
 
# read a titanic.csv file
# from seaborn library
df = sns.load_dataset('titanic')
 
# class v / s fare barplot
sns.barplot(x = 'class', y = 'fare', data = df)
 
# Show the plot
plt.show()
```

# Draw a set of vertical bars with nested grouping by two variables. 

## Creating a bar plot using hue parameter with two variable

```python
# from seaborn library
df = sns.load_dataset('titanic')
 
# class v / s fare barplot
sns.barplot(x = 'class', y = 'fare', hue = 'sex', data = df)
 
# Show the plot
plt.show()
```

# shows a Horizontal barplot. 

```python
# from seaborn library
df = sns.load_dataset('titanic')
 
# fare v / s class horizontal barplot
sns.barplot(x = 'fare', y = 'class', hue = 'sex', data = df)
 
# Show the plot
plt.show()
```

# Plot all bars in a given order 

```python
# from seaborn library
df = sns.load_dataset('titanic')
 
# class v / s fare barplot in given order
sns.barplot(x = 'class', y = 'fare', data = df,
            order = ["Third", "Second", "First"])
 
# Show the plot
plt.show()
```

# Plot all bars in a single color using color attributes. 

```python
# read a titanic.csv file from seaborn library
df = sns.load_dataset('titanic')
 
# class v / s fare barplot with same colour
sns.barplot(x = 'class', y = 'fare', data = df, color = "red")
 
# Show the plot
plt.show()
```

# barplot without error bars using ci attributes. 

```python
df = sns.load_dataset('titanic')
 
# class v / s fare barplot
# without error bars
sns.barplot(x = 'class', y = 'fare', data = df, ci = None)
 
# Show the plot
plt.show()
```

# Colors to use for the different levels of the hue variable using palette.

```python
df = sns.load_dataset('titanic')
 
sns.barplot(x = 'class', y = 'fare',
            hue = 'sex', data = df, palette='Set2')
 
# Show the plot
plt.show()
```

# Using the statistical function NumPy.median and NumPy.mean to estimate within each categorical bin.

```python
# read a titanic.csv file
# from seaborn library
from numpy import median
df = sns.load_dataset('titanic')
 
sns.barplot(x = 'class', y = 'fare', hue = 'sex', data = df, estimator=median)
#sns.barplot(x = 'class', y = 'fare', hue = 'sex', data = df, estimator=np.mean) 
# Show the plot
plt.show()
```

# Using the saturation parameter

```python
df = sns.load_dataset('titanic')
 
sns.barplot(x = 'class', y = 'fare', hue = 'sex', data = df,saturation = 0.1)
 
# Show the plot
plt.show()
```

# Use matplotlib.axes.Axes.bar() parameters to control the style.

```python
df = sns.load_dataset('titanic')
 
sns.barplot(x="class", y="fare", data=df,
                 linewidth=2.9, facecolor=(1, 1, 1, 0),
                 errcolor=".2", edgecolor=".2")
```
